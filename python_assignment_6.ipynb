```python
#Question 2 : List down the Benefits of OOP?

# Answer:
# DRY code (Don't repeat yourself)
# Eaiser maintaince and modification
# Fewier errors
# Modularity (same classes and methods can be used in other programs)
```

```python
#Question 3 : Differentiate between function and method?

# Answer:
# A function is a block of code that is defined in the global scope of a program, that can be also invoked by other
# functions, methods or events

# A method is function that is defined inside of a class that is only available to that class and its instances
```

```python
#Question 4 : Define the following terms: 1. Class 2. Object 3. Attribute 4. Behavior

#Answer:

#1. Class
# A class can be defined as a blueprint for an object for example if a class has defined the variable x,y,z and the method fun()
# the Objects created from this class will inhert the variable x,y,z and the method fun()

#2. Object
# An Object is an instance of class, A object can have methods and properties. there can be multiple objects of the same class
# different names

#3. Attribute
# Attribute can be defined as the properties of a Object. for example in the object [person.name] person is the object and name
# is its attribute

#4. Behavier
# The term Behavier can be defined as the output or the result the we expect from a program or from a block of code. when
# we create an instance of a class we expect it to inhert all the attributes and the methods of the class and when those methods
# are called it should behave exactly as we have define in the blueprint class
```

```python
#Question 5 : 
# Write a code in python in which create a class named it Car which
# have 5 attributes such like (model, color and name etc.) and 3
# methods. And create 5 object instance from that class.

class Car:
    def __init__(self,color,brand,top_speed,year_made,price):
        self.color = color
        self.brand = brand
        self.top_speed = top_speed
        self.year_made = year_made
        self.price = price

    def get_info(self):
        message = f'Details:\nCar Brand: {self.brand}\nCar Color: {self.color}\nCar Top speed: {self.top_speed}\nCar Year Made: {self.year_made}\nCar Price: {self.price}\n'
        print(message)

    def resale_value(self):
        resale_value = self.price * 0.8
        print(f'Resale value of this car is {resale_value}\n')

    def repair_cost(self):
        repair_cost = self.price * 0.03
        print(f'This car will cost you {repair_cost} per year in repairs\n')


Honda_civic = Car('Red','Honda',120,2011,1200000)
Toyota_vitz = Car('white', 'Toyota', 270, 2017, 2400000)
Suzuki_swift = Car('silver', 'Suzuki', 245, 2018, 2350000)
Mazda_rx8 = Car('yellow', 'Mazda', 345, 2019, 3250000)
Tesla_p100 = Car('black', 'Tesla', 390, 2019, 4545000)

cars = [Honda_civic, Toyota_vitz, Tesla_p100, Suzuki_swift, Mazda_rx8]
for car in cars:
    car.get_info()
    car.resale_value()
    car.repair_cost()
    print('################################################')
        
```
